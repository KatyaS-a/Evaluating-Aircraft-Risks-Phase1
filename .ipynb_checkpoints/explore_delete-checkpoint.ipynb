```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Custom date parser functions
date_parser_1 = lambda x: pd.to_datetime(x, format='%Y-%m-%d')

df = pd.read_csv('data/AviationData_Grouped.csv', encoding='ISO-8859-1', converters={'Event.Date': date_parser_1}, index_col=0, low_memory=False)
df.head()
```

```python
df.info()
```

```python
# Extract year from Event.Date
df['Year'] = df['Event.Date'].dt.year
```

```python
df['Year'].value_counts().sort_index()
```

```python
# Group by year and sum the injury columns
result = df.groupby('Year').agg({
    'Total.Fatal.Injuries': 'sum',
    'Total.Serious.Injuries': 'sum',
    'Total.Minor.Injuries': 'sum',
    'Total.Uninjured': 'sum'
}).reset_index()

print(result)
```
