```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('data/AviationData_Grouped.csv', index_col=0, low_memory=False)
df.head()
```

```python
# Replace values in 'Purpose.of.flight'
df['Purpose.of.flight'] = df['Purpose.of.flight'].replace(['Public Aircraft - Local', 
                                                           'Public Aircraft - State', 'Public Aircraft - Federal'], 'Public Aircraft')
```

```python
df.shape
```

```python
df['Event.Date'].value_counts().sort_index()
```

```python
# Remove instances before 1980
df = df[df['Event.Date'] >= '1980-01-01']
```

```python
# Calculate the percentage of missing values in each column
(df.isnull().sum() * 100 / len(df)).round(2)
```

```python
df.shape
```

```python
mm_unique = (df['General_Maker_Model'].unique()).tolist()
mm_unique = sorted(mm_unique)
#mm_unique = ', '.join(mm_unique)
display(len(mm_unique))
```

```python
# Aircraft that are no longer in use
vintage = ['aero_commander_100', 'aero_commander_500', 'aero_commander_S2', 'aeronca_11', 'aeronca_15', 
           'aeronca_7', 'alon_a2', 'bellanca_1730', 'bellanca_1731', 'bellanca_7', 'bellanca_8', 'boeing_a75', 
           'boeing_b75', 'boeing_e75', 'globe_gc1', 'great_lakes_2t1', 'republic_rc3', 'ryan_st3', 
           'schweizer_sgs134', 'schweizer_sgs233', 'taylorcraft_bc12d', 'taylorcraft_bl65', 'taylorcraft_f19', 
           'waco_upf7']

vintage_df = df[df['General_Maker_Model'].isin(vintage)]

vintage_df.shape
```

```python
# Keep aircraft that are still in use in df
df = df[~df['General_Maker_Model'].isin(vintage)]

df.shape
```

```python
df.sort_values(by='Total.Fatal.Injuries', ascending=False)
```

```python
# Calculate the number of people on board each flight
df['Total_On_Board'] = (df['Total.Fatal.Injuries'] + 
                        df['Total.Serious.Injuries'] + 
                        df['Total.Minor.Injuries'] + 
                        df['Total.Uninjured'])
```

```python
df.head()
```

```python
df['General_Maker_Model'].value_counts()
```

```python
# Set display options to show all rows
pd.set_option('display.max_rows', None)

df['Total_On_Board'].value_counts().sort_index()
```

```python
df[df['Total_On_Board'] == 0]
```

```python
# Remove instances where the aircraft was unoccupied when an accident happened
df = df[df['Total_On_Board'] != 0]
```

```python
# Save the DataFrame to a CSV file just in case
df.to_csv('Aviation_from_Visualizationa.csv', index=True)
```

```python
# Step 1: Calculate the total number of instances for each General_Maker_Model
total_counts = df['General_Maker_Model'].value_counts()
total_counts.sort_index()
```

```python
# Step 2: Calculate the number of instances where Aircraft.damage is "Destroyed" for each General_Maker_Model
destroyed_counts = df[df['Aircraft.damage'] == 'Destroyed']['General_Maker_Model'].value_counts()
destroyed_counts.sort_index()
```

```python
# Step 3: Compute the percentage
percentage_destroyed = (destroyed_counts / total_counts) * 100
percentage_destroyed.sort_values()
```

```python
# Calculate the percentage of 'Destroyed' aircraft for each General_Maker_Model
damage_counts = df.groupby('General_Maker_Model')['Aircraft.damage'].value_counts(normalize=True).unstack().fillna(0)
damage_counts['Destroyed_Percentage'] = damage_counts['Destroyed'] * 100

# Sort by Destroyed_Percentage and select the top 10
top_10_damage_counts = damage_counts.sort_values(by='Destroyed_Percentage', ascending=False).head(10)

# Plot the data
plt.figure(figsize=(10, 6))
plt.bar(top_10_damage_counts.index, top_10_damage_counts['Destroyed_Percentage'], color='royalblue')
plt.xlabel('Maker_Model')
plt.ylabel('Percentage of Destroyed Aircraft')
plt.title('Top 10 Aircraft Models by Percentage of Destroyed Aircraft')
plt.xticks(rotation=60)
plt.show()
```

```python
# Calculate the proportion of Total.Fatal.Injuries out of Total_On_Board
df['Proportion_Fatal'] = df['Total.Fatal.Injuries'] / df['Total_On_Board']

# Group by General_Maker_Model and calculate the mean proportion
grouped_df = df.groupby('General_Maker_Model')['Proportion_Fatal'].mean().reset_index()

# Sort by Proportion_Fatal and select the top 10
top_10_df = grouped_df.sort_values(by='Proportion_Fatal', ascending=False).head(10)

# Plot the data
plt.figure(figsize=(10, 6))
plt.bar(top_10_df['General_Maker_Model'], top_10_df['Proportion_Fatal'], color='royalblue')
plt.xlabel('Maker_Model')
plt.ylabel('Proportion of Fatal Injuries')
plt.title('Top 10 Aircraft Models by Proportion of Fatalities')
plt.xticks(rotation=80)
plt.show()
```

```python
# Calculate the proportion of Total.Fatal.Injuries out of Total_On_Board
df['Proportion_Fatal'] = df['Total.Fatal.Injuries'] / df['Total_On_Board']

# Group by General_Maker_Model and calculate the mean proportion
grouped_df = df.groupby('General_Maker_Model')['Proportion_Fatal'].mean().reset_index()

# Sort by Proportion_Fatal and select the top 10
top_10_df = grouped_df.sort_values(by='Proportion_Fatal', ascending=False).tail(10)

# Plot the data
plt.figure(figsize=(10, 6))
plt.bar(top_10_df['General_Maker_Model'], top_10_df['Proportion_Fatal'], color='royalblue')
plt.xlabel('General Maker Model')
plt.ylabel('Proportion of Fatal Injuries')
plt.title('Bottom 10 General Maker Models by Proportion of Fatal Injuries')
plt.xticks(rotation=80)
plt.show()
```
